Para encontrar una solución numérica a la integral:

$\int_{0}^{a}{\frac{da}{\sqrt{\Omega_{r,0}/a^{2}+\Omega_{m,0}/a+\Omega_{\Lambda,0}a^{2}+(1-\Omega_{0})}}}=H_{0}t$

usaremos valores para las densidades reportadas en https://arxiv.org/abs/1807.06209 (Planck,2018):

$\Omega_{m,0}=0.315$,$\Omega_{\Lambda,0}=0.6847$,$\Omega_{r,0}=0.24$ $H_{0}=67.4 km/s Mpc$

Donde consideramos la materia oscura y la materia bariónica en $\Omega_{m,0}$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def trapecio(omega_0,a,b,n):
    y = 0
    h = (b-a)/n
    x = np.linspace(0,b,n)
    for i in range(n):
        y += 1/np.sqrt((0.24/(a+(i*h)))+(0.315/(a+(i*h)))+(0.648*(a+(i*h)))+(1-omega_0))
    return y*(h/2)
```

```python
def simpson(omega_0,a,b):
    y = 1/(np.sqrt(omega_0/a+(1-omega_0)))
    I = ((b-a)/6)*((1/(np.sqrt(omega_0/a+(1-omega_0))))+4*(1/(np.sqrt(omega_0/(a+b)+(1-omega_0))))+1/(np.sqrt(omega_0/b+(1-omega_0))))
    return I
```

```python
I = simpson(1,0.01,1)
print(I)
```

```python
r = np.linspace(0.01,50,1000)
I = np.zeros(1000)
for i in range(len(r)):
    I[i] = simpson(1,0.01,r[i])
print(I)    
```

```python
r = np.linspace(0.01,50,1000)
I_2 = np.zeros(1000)
for i in range(len(r)):
    I_2[i] = simpson(0.9,0.01,r[i])
print(I_2) 
```

```python
r = np.linspace(0.01,50,1000)
I_3 = np.zeros(1000)
for i in range(len(r)):
    I_3[i] = simpson(1.1,0.01,r[i])
print(I_3) 
```

```python
plt.plot(I, r, color = 'green', linestyle = '--', label = '$\Omega_{0}=1$')
plt.plot(I_2,r, color = 'black', label = '$\Omega_{0}=0.9$')
plt.plot(I_3,r, color = 'gray', linestyle = ':', label = '$\Omega_{0}=1.1$')
plt.xlim(-0.05,120)
plt.ylim(0,50)
plt.xlabel('$H_{0}t$')
plt.ylabel('a')
plt.grid()
plt.legend()
plt.savefig('omega_0_behaviour')
```
