```python
import numpy as np
import matplotlib.pyplot as plt
```

The energy densities as functions of $a$:

$\epsilon_{m}\propto a^{-3}$

$\epsilon_{\gamma}\propto a^{-4}$

$\epsilon_{\Lambda}\propto 1$

as functions of $z$:

$\epsilon_{m}\propto (1+z)^{3}$

$\epsilon_{\gamma}\propto (1+z)^{4}$

$\epsilon_{\Lambda}\propto 1$

```python
a = np.linspace(0.1, 10, 1000) #Defining the range of scale factors
```

```python
#Plot for the scale factor
plt.plot(a,a**-3,color ='green', linestyle = '-', label = '$\epsilon_{M}\propto a^{-3}$')
plt.plot(a,a**-4,color ='black', linestyle = '--', label = '$\epsilon_{\gamma}\propto a^{-4}$')
plt.plot(a,np.ones(1000), color ='black', linestyle = ':',label = '$\epsilon_{\Lambda} = constant$')
plt.legend()
plt.grid()
plt.xlabel('$a$')
plt.ylabel('$\epsilon$')
plt.title('Energy densities as functions of the scale factor ($a$)')
plt.xlim(0.1,2)
plt.ylim(0,10)
plt.savefig('Energy_densities_as_func_scale_f')
```

```python
#Plot for the redshit
plt.plot(1/a-1,(1/a-1)**3,color ='green', linestyle = '-', label = '$\epsilon_{M}\propto (1+z)^{3}$')
plt.plot(1/a-1,(1/a-1)**4,color ='black', linestyle = '--', label = '$\epsilon_{\gamma}\propto (1+z)^{4}$')
plt.plot(1/a-1,np.ones(1000), color ='black', linestyle = ':',label = '$\epsilon_{\Lambda} = constant$')
plt.legend()
plt.grid()
plt.xlabel('$z$')
plt.ylabel('$\epsilon$')
plt.title('Energy densities as functions of the redshift ($z$)')
plt.xlim(0.1,2)
plt.ylim(0,10)
plt.savefig('Energy_densities_as_func_redshift')
```
