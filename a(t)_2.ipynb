```python
import numpy as np
import matplotlib.pyplot as plt
```

$a(t)\approx t^{1/2}$

$a(t)\approx t^{2/3}$

$a(t) = e^{H_{0}\Omega_{DE,0}^{1/2}t}$

```python
t = np.linspace(0,1,100)
H_0=70
plt.plot(t,t**1/2, color = 'green', linestyle = '--', label = '$a(t) \propto t^{1/2}$')
#plt.plot(t,t**2/3, color = 'black', linestyle = '-', label = '$a(t)\propto t^{3/2}$')
#plt.plot(t,np.exp(70*(0.7**1/2)*t), color = 'gray', linestyle = ':', label = '$a(t) = e^{H_{0}\Omega_{DE,0}^{1/2}t}$')
plt.xlabel('t')
plt.ylabel('a')
plt.title('Scale factor for radiation only')
plt.legend()
plt.grid()
```

```python
t = np.linspace(0,1,100)
H_0=70
#plt.plot(t,t**1/2, color = 'green', linestyle = '--', label = '$a(t) \propto t^{1/2}$')
plt.plot(t,t**2/3, color = 'black', linestyle = '-', label = '$a(t)\propto t^{3/2}$')
#plt.plot(t,np.exp(70*(0.7**1/2)*t), color = 'gray', linestyle = ':', label = '$a(t) = e^{H_{0}\Omega_{DE,0}^{1/2}t}$')
plt.xlabel('t')
plt.ylabel('a')
plt.title('Scale factor for matter only')
plt.legend()
plt.grid()
```

```python
t = np.linspace(0,1,100)
H_0=70
#plt.plot(t,t**1/2, color = 'green', linestyle = '--', label = '$a(t) \propto t^{1/2}$')
#plt.plot(t,t**2/3, color = 'black', linestyle = '-', label = '$a(t)\propto t^{3/2}$')
plt.plot(t,np.exp(70*(0.7**1/2)*t), color = 'gray', linestyle = ':', label = '$a(t) = e^{H_{0}\Omega_{DE,0}^{1/2}t}$')
plt.xlabel('t')
plt.ylabel('a')
plt.title('Scale factor for dark energy only')
plt.legend()
plt.grid()
```

```python
t = np.linspace(0,10,100)
H_0=70
plt.plot(t,t**1/2, color = 'green', linestyle = '--', label = '$a(t) \propto t^{1/2}$')
plt.plot(t,t**2/3, color = 'black', linestyle = '-', label = '$a(t)\propto t^{3/2}$')
plt.plot(t,np.exp(70*(np.sqrt(0.7))*(t)), color = 'blue', linestyle = ':', label = '$a(t) = e^{H_{0}\Omega_{DE,0}^{1/2}t}$')
plt.xlabel('t')
plt.ylabel('a')
plt.xlim(-0.05,2)
plt.ylim(0,1.5)
plt.title('')
plt.legend()
plt.grid()
plt.savefig('a(t)_analytic_sol')
```
